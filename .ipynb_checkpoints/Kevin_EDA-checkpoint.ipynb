# **EDA Used for Additional Insights**

```python
# Import basic libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#set plot style to 'ggplot' and don't limit the view of DF when diaplyhign them to screen
plt.style.use('ggplot')
pd.options.display.max_columns = None
%matplotlib inline
```

```python
#importing our data and resetting our index
df = pd.read_excel('Data/Comp2_Raw_Data.xls')
df.head(5)
```

```python
df = df.drop(columns = ['ID'])
df.columns = ['Credit_Limit', 'Gender', 'Education', 'Marriage',  'Age', 'Pay_Sept', 'Pay_Aug', 'Pay_Jul', 'Pay_Jun', 'Pay_May', 'Pay_Apr', 
             'Bill_Amt_Sept', 'Bill_Amt_Aug', 'Bill_Amt_Jul', 'Bill_Amt_Jun', 'Bill_Amt_May', 'Bill_Amt_Apr','Pay_Amt_Sept', 'Pay_Amt_Aug',
             'Pay_Amt_Jul', 'Pay_Amt_Jun', 'Pay_Amt_May', 'Pay_Amt_Apr', 'Default']
df.head(5)
```

```python
#Probability of Defaulting
default_prob = df['Default'].sum() / len(df['Default'])
print('The probability of defaulting is', default_prob)
```

```python
pd.crosstab(df.Gender, df.Default)
```

```python
#Probability of Defaulting by Gender
prob_default_gender = df[['Gender', "Default"]].groupby(['Gender'],as_index=False).mean().sort_values(by='Default', ascending=False)
                                                      
print(prob_default_gender)
```

```python
df['Gender'].value_counts()
```

```python
#Percentage of Women in the Dataset
percent_women = df.Gender.value_counts() [2] / len(df.Gender)
print('The percentage of women in the dataset is', percent_women)

#Percentage of Men in the Dataset
percent_men = df.Gender.value_counts() [1] / len(df.Gender)
print('The percentage of men in the dataset is', percent_men)
```

```python
#Percentage of Married People in the Dataset
percent_married = df.Marriage.value_counts() [1] / len(df.Marriage)
print('The percentage of married people in the dataset is', percent_married)

#By Gender
prob_married_gender = df[['Gender', "Marriage"]].groupby(['Gender'],as_index=False).mean().sort_values(by='Marriage', ascending=False)
print(prob_married_gender)

#Percentage of College Educated in the Dataset
percent_college = df.Education.value_counts() [2] / len(df.Education)
print('The percentage of people with a college education in the dataset is', percent_college)

#By Gender
prob_college_gender = df[['Gender', "Education"]].groupby(['Gender'],as_index=False).mean().sort_values(by='Education', ascending=False)
print(prob_college_gender)
```

```python
#More Info On Defaulting

#By Marriage
prob_married_default = df[['Default', "Marriage"]].groupby(['Default'],as_index=False).mean().sort_values(by='Marriage', ascending=False)
print(prob_married_default)

#By College Education
prob_college_gender = df[['Gender', "Education"]].groupby(['Gender'],as_index=False).mean().sort_values(by='Education', ascending=False)
print(prob_college_gender)

```

```python
df['Age_bins'] = np.where(df.Age <=df.Age.quantile(.25),1,
                          np.where(df.Age >df.Age.quantile(.25)<=df.Age.quantile(.50),2,
                          np.where(df.Age>df.Age.quantile(.50)<=df.Age.quantile(.75),3,
                          np.where(df.Age>df.Age.quantile(.75),4,5))))








```

```python
def assign_age_bins(Age):
    if df.Age <= df.Age.quantile(.25):                    
        return(1)
    if df.Age > df.Age.quantile(.25) & df.Age <= df.Age.quantile(.5):        
        return(2)
    if df.Age > df.Age.quantile(.5) & df.Age <= df.Age.quantile(.75):       
        return(3)
    else:      
        return(4)

df['age_bins'] = df['Age'].apply(assign_age_bins)  

df.head(5)
```
