```python
df['Credit_Utilization_Sept'] = df['Bill_Amt_Sept'] / df['Credit_Limit']
df['Credit_Utilization_Aug'] = df['Bill_Amt_Aug'] / df['Credit_Limit']
df['Credit_Utilization_Jul'] = df['Bill_Amt_Jul'] / df['Credit_Limit']
df['Credit_Utilization_Jun'] = df['Bill_Amt_Jun'] / df['Credit_Limit']
df['Credit_Utilization_May'] = df['Bill_Amt_May'] / df['Credit_Limit']
df['Credit_Utilization_Apr'] = df['Bill_Amt_Apr'] / df['Credit_Limit']
df['Total_Billed_Amt'] = df['Bill_Amt_Sept'] + df['Bill_Amt_Aug'] + df['Bill_Amt_Jul'] + df['Bill_Amt_Jun'] + df['Bill_Amt_May'] + df['Bill_Amt_Apr']
df['Total_Pay_Amt'] = df['Pay_Amt_Sept'] + df['Pay_Amt_Aug'] + df['Pay_Amt_Jul'] + df['Pay_Amt_Jun'] + df['Pay_Amt_May'] + df['Pay_Amt_Apr']
df['Outstanding_Debt'] = df['Total_Billed_Amt'] - df['Total_Pay_Amt']
df['Over_Limit_Sept'] = np.where(df['Credit_Utilization_Sept']>1,1,0)
df['Over_Limit_Aug'] = np.where(df['Credit_Utilization_Aug']>1,1,0)
df['Over_Limit_Jul'] = np.where(df['Credit_Utilization_Jul']>1,1,0)
df['Over_Limit_Jun'] = np.where(df['Credit_Utilization_Jun']>1,1,0)
df['Over_Limit_May'] = np.where(df['Credit_Utilization_May']>1,1,0)
df['Over_Limit_Apr'] = np.where(df['Credit_Utilization_Apr']>1,1,0)
```

```python
df['Binned_Credit_Limit'] = pd.qcut(df['Credit_Limit'], q=4, labels=[0,1,2,3])
df['Credit_Limit0'] = np.where(df['Binned_Credit_Limit']==0,1,0)
df['Credit_Limit1'] = np.where(df['Binned_Credit_Limit']==1,1,0)
df['Credit_Limit2'] = np.where(df['Binned_Credit_Limit']==2,1,0)
df['Credit_Limit3'] = np.where(df['Binned_Credit_Limit']==3,1,0)
```

```python
df['Gender_M'] = np.where(df['Gender'] ==1, 1, 0)
df['Gender_F'] = np.where(df['Gender'] ==2, 1, 0)
```

```python
df['Binned_Age'] = pd.qcut(df['Age'], q=4, labels=[0,1,2,3])
df['Age_Range0'] = np.where(df['Binned_Age'] ==0, 1, 0)
df['Age_Range1'] = np.where(df['Binned_Age'] ==1, 1, 0)
df['Age_Range2'] = np.where(df['Binned_Age'] ==2, 1, 0)
df['Age_Range3'] = np.where(df['Binned_Age'] ==3, 1, 0)
```

```python
df['Education'] = np.where(df['Education'] == 2,1,0)
df['Edu_College'] = np.where(df['Education']==1,1,0)
df['Edu_Other'] = np.where(df['Education']==0,1,0)
```

```python
df['Marriage'] = np.where(df['Marriage']==1,1,0)
df['Married'] = np.where(df['Marriage'] ==1, 1, 0)
df['Married_Other'] = np.where(df['Marriage']==0,1,0)
```

```python
df = df.drop(columns = ['Credit_Limit', 'Gender', 'Education', 'Marriage','Age','Binned_Credit_Limit', 'Binned_Age'])
```
